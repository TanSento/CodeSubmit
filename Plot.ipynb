# Plotting Train Set and Test Set

```python
def get_data(name):
  # returns a T x 3 list of stock prices
  # each row is a different stock
  # 0 = AAPL
  # 1 = MSI
  # 2 = SBUX
  df = pd.read_csv(name)
  return df.values    # get a numpy array
                      # each row corresponds to one day
```

```python
train_1st = get_data("UltimateStocks.csv")[:-440]
test_1st = get_data("UltimateStocks.csv")[-440:]

train_2nd = get_data("train_2nd.csv")
test_2nd = get_data("test_2nd.csv")

train_3rd = get_data("train_3rd.csv")
test_3rd = get_data("test_3rd.csv")

train_4th = get_data("train_4th.csv")
test_4th = get_data("test_4th.csv")
```

**1st Split**

```python
stock1 = train_1st[:,0]
stock2 = train_1st[:,1]
stock3 = train_1st[:,2]

figure(figsize=(8, 6), dpi=80)
plt.plot(stock1, label = "Apple")      
plt.plot(stock2, label = "Motorola")   
plt.plot(stock3, label = "Starbucks")   
plt.legend()
plt.title("Train Set 1st")
plt.grid()
plt.show()
```

```python
stock1 = test_1st[:,0]
stock2 = test_1st[:,1]
stock3 = test_1st[:,2]

figure(figsize=(8, 6), dpi=80)
plt.plot(stock1, label = "Apple")      
plt.plot(stock2, label = "Motorola")   
plt.plot(stock3, label = "Starbucks") 
plt.title("Test Set 1st")
plt.grid()
plt.legend()
plt.show()
```
